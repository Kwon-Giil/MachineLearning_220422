 - 다중분류: iris, 배스트모델, 조기종료

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

 - 데이터 전처리

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
iris_std = StandardScaler().fit_transform(iris.data)
y_onehot = to_categorical(iris.target)
X_train, X_test, Y_train, Y_test = train_test_split(iris_std, y_onehot, stratify=y_onehot, test_size=0.2, random_state=seed)
```

 - 모델 정의 / 설정 / 학습 / 평가

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
model = Sequential([Dense(12, input_dim=4, activation='relu'),
                    Dense(8, activation='relu'),
                    Dense(3,activation='softmax')])
model.summary()
```

```python
model.compile('adam','categorical_crossentropy', metrics=['accuracy'])
```

```python
mc = ModelCheckpoint('iris_best.h5', monitor='val_loss', verbose=1, save_best_only=True)
es = EarlyStopping(patience = 30)
```

```python
hist = model.fit(X_train, Y_train, validation_split=0.2, epochs=1000, batch_size=100, verbose=0, callbacks= [mc,es])
```

```python
model.evaluate(X_test, Y_test)
```

```python
best_model = load_model('iris_best.h5')
best_model.evaluate(X_test, Y_test)
```

```python
y_acc = hist.history['accuracy']
y_vloss = hist.history['val_loss']
xs = np.arange(1,len(y_acc)+1)
```

```python
plt.figure(figsize=(12,8))
plt.plot(xs, y_acc, label='accuracy')
plt.plot(xs, y_vloss, label='val_loss')
plt.legend(), plt.grid()
plt.ylim([0,1.2])
plt.xlabel('epoch')
plt.show()
```
