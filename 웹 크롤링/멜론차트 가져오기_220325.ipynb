```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import time
import tqdm
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
time.sleep(2)
url = 'https://www.melon.com/chart/day/index.htm'
driver.get(url)
driver.implicitly_wait(10)
# url 가져오는 거

html = driver.page_source
soup = bs(html, 'html.parser')

tbody_soup = soup.find('tbody')
tr_soup = tbody_soup.find_all('tr')

print(tr_soup[0].find('span',class_='rank').get_text()) # 곡 순위
print(tr_soup[0].find('div',class_='ellipsis rank01').get_text()) # 곡 이름
print(tr_soup[0].find('span',class_='checkEllipsis').get_text()) # 아티스트 이름
print(tr_soup[0].find('div',class_='ellipsis rank03').get_text()) # 앨범 명
```

```python
# 멜론 순위 전체 데이터 가져온 뒤에 DataFrame으로 출력하기
from tqdm import tqdm
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
time.sleep(2)
url = 'https://www.melon.com/chart/day/index.htm'
driver.get(url)
time.sleep(2)

html = driver.page_source
soup = bs(html, 'html.parser')
driver.quit()

tbody_soup = soup.find('tbody')
tr_soup = tbody_soup.find_all('tr')

ml_day = []

for tr in tqdm(tr_soup):
    rank=tr.find('span',class_='rank').get_text().replace('\n','')
    title=tr.find('div',class_='ellipsis rank01').get_text().replace('\n','')
    artist=tr.find('span',class_='checkEllipsis').get_text().replace('\n','')
    album=tr.find('div',class_='ellipsis rank03').get_text().replace('\n','')
    
    ml_day.append([rank,title,artist,album])
    
df = pd.DataFrame(ml_day,columns=['순위','제목','아티스트','앨범명'])
df.reset_index(drop=True, inplace =True)
df
```
