```python
# 네이버 영화 페이지 접속
url='https://movie.naver.com/movie/point/af/list.naver'
driver=webdriver.Chrome('./chromedriver.exe')
driver.get(url) # 네이버 영화 평점의 첫 페이지를 불러오는 거
html = driver.page_source
soup = bs(html,'html.parser')

# 네이버 영화 리뷰 데이터 추출하기
r1_t = []
r1_d = []
rating= []
titles = []

for page in tqdm(range(1,102)): # 페이지를 바꿔주는 거
    driver.get(url+'?&page='+str(page)) # 각 페이지를 접속한다
    html = bs(driver.page_source, 'html.parser')
    soup_tr1=html.find_all('tr')
    time.sleep(1)

    for i in range(1,10): # 각 페이지의 댓글 숫자만큼 긁어오기
        title = soup_tr[i].find('a', class_= 'movie color_b').get_text()

        score = soup_tr[i].find('em').get_text()

        soup_a = soup_tr[i].find_all('a',class_='report')
        rtxt=str(soup_a).split(',')[2]

        soup_date = soup_tr[i].find_all('td',class_='num')
        rdate=str(soup_date).split(',')[-1][-14:-6]

        titles.append(title) # 영화 제목만 따로 저장
        rating.append(score) # 평점을 저장
        r1_t.append(rtxt) # 댓글을 저장
        r1_d.append(rdate) # 댓글 작성 날짜를 저장

# 4개의 리스트를 각각의 필드로 만들어서 데이터프레임으로 출력
df_1p = pd.DataFrame(zip(titles,r1_t,rating,r1_d), columns = ['제목','댓글','평점','작성일자'])
df_1p
```

```python
# 데이터 저장
df_1p.to_csv('naver_movie_review.csv', index = False)
```

```python
# TextBlob을 이용한 감정 분석
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier
```

```python
train_data = pd.read_table('./data/ratings_train.txt')
train_data
```

```python
# 훈련용 데이터 빼오기
pos_df = train_data[train_data['label']==1].head(1000)
neg_df = train_data[train_data['label']==0].head(1000)
tr_data1 = pd.concat([pos_df, neg_df])

tr_data1=tr_data1.sort_values('id')
tr_data1 = tr_data1.reset_index(drop=True)
tr_data1
```

```python
# NaN 확인
tr_data1.isna().sum()
```

```python
# 테스트용 데이터 빼오기
pos_df1 = train_data[train_data['label']==1].tail(250)
neg_df1 = train_data[train_data['label']==0].tail(250)
test_data = pd.concat([pos_df1, neg_df1])

test_data=test_data.sort_values('id')
test_data = test_data.reset_index(drop=True)


test_data.to_csv('ratings_test.txt')
```

```python
# NaN 확인
test_data.isna().sum()
```

```python
# train , test 데이터 리뷰의 id 제거
del tr_data1['id'], test_data['id']
```

```python
# 튜플 구조로 변경하기
def tup_che(df_t):
    tr_te =[]
    for i in range(len(df_t)):
        tr_te.append((df_t.iloc[i, 0], df_t.iloc[i, 1]))
    return tr_te
```

```python
train = tup_che(tr_data1)
test = tup_che(test_data)

train[:3]
```

```python
pos_neg = NaiveBayesClassifier(train)
test_acc = pos_neg.accuracy(test)

print(f'정확도: {test_acc}')
```
