```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
```

```python
# 드라이버 생성하여 url 가져오기
driver = webdriver.Chrome('C:/Users/kwoni/Desktop/pycode/chromedriver.exe')
url='https://section.blog.naver.com/Search/Post.naver?pageNo=1&rangeType=ALL&orderBy=sim&keyword=%EA%B5%AD%EB%82%B4%20%EC%97%AC%ED%96%89%EC%A7%80'
driver.get(url)
time.sleep(2) #2초 정도 의도적으로 딜레이 부여

html=driver.page_source
# requests에서 사용하는 방법과는 다름
#find_all 을 쓰기는 힘들다.

soup = bs(html,'html.parser')
soup.find_all('div',class_='list_search_post')

# 나중에 클릭할 때
#txt_soup=driver.find_elements_by_css_selector('title') # 검색 데이터 제목 찾아오는 기능
#txt_soup
```

```python
post_soup = soup.find_all('div',class_='list_search_post')
post_soup[0].find('strong',class_='title_post').get_text()
print(post_soup[0].find('a')['href'])
```

```python
# 검색화면 제목만 뽑아내는 기능
post_soup = soup.find_all('div',class_='list_search_post')
post_list=[]
for post in post_soup:
    title=post.find('strong',class_='title_post').get_text()
    href=post.find('a')['href']
    post_list.append({'제목': title, '링크':href})
    
df = pd.DataFrame(post_list)
df
```

```python
for i in df.index:
    driver.get(df.loc[i,'링크'])
    time.sleep(2)
    
    html=driver.page_source
    soup=bs(html,'html.parser')
    print("http://blog.naver.com"+soup.find('iframe',id='mainFrame')['src'])
# 검색 결과의 링크들만 뽑아내는 기능
```

```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By # 자동으로 데이터를 구현해서 사용 가능
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import pyperclip
import time
```

```python
# 크롬 드라이브 실행 및 네이버 이동
driver = Chrome('./chromedriver.exe')
driver.get('http://www.naver.com')
```
